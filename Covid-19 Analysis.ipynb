```r
covid_data$Last_Update <- as.Date(covid_data$Last_Update)
# Trim down the time for consistency


covid_data_dailySum <- covid_data %>%
    group_by(Last_Update) %>% 
    summarise(Total_Confirmed = sum(Confirmed), Deaths = sum(Deaths), Mean_CFRatio = mean(Case_Fatality_Ratio, na.rm=TRUE))


covid_data_dailySum$type <-  covid_data_dailySum$Deaths / covid_data_dailySum$Deaths
covid_data_dailySum
```

```r
covid_data_east$Last_Update <- as.Date(covid_data_east$Last_Update)
# Trim down the time for consistency


covid_data_dailySum_east <- covid_data_east %>%
    group_by(Last_Update) %>% 
    summarise(Total_Confirmed = sum(Confirmed), Deaths = sum(Deaths), Mean_CFRatio = mean(Case_Fatality_Ratio, na.rm=TRUE))

covid_data_dailySum_east$type <-  covid_data_dailySum_east$Deaths / covid_data_dailySum_east$Deaths*2
covid_data_dailySum_east
```

```r
covid_data_west$Last_Update <- as.Date(covid_data_west$Last_Update)
# Trim down the time for consistency


covid_data_dailySum_west <- covid_data_west %>%
    group_by(Last_Update) %>% 
    summarise(Total_Confirmed = sum(Confirmed), Deaths = sum(Deaths), Mean_CFRatio = mean(Case_Fatality_Ratio, na.rm=TRUE))

covid_data_dailySum_west$type <-  covid_data_dailySum_west$Deaths / covid_data_dailySum_west$Deaths*3
covid_data_dailySum_west
```

```r
covid_data_mid$Last_Update <- as.Date(covid_data_mid$Last_Update)
# Trim down the time for consistency


covid_data_dailySum_mid <- covid_data_mid %>%
    group_by(Last_Update) %>% 
    summarise(Total_Confirmed = sum(Confirmed), Deaths = sum(Deaths), Mean_CFRatio = mean(Case_Fatality_Ratio, na.rm=TRUE))

covid_data_dailySum_mid$type <-  covid_data_dailySum_mid$Deaths / covid_data_dailySum_mid$Deaths*4
covid_data_dailySum_mid
```

```r
covid_final<-do.call("rbind", list(covid_data_dailySum, covid_data_dailySum_east, covid_data_dailySum_west,covid_data_dailySum_mid))
covid_final <- covid_final %>%
    mutate(type=as_factor(type))
covid_final
```

```r
options(repr.plot.width=20, repr.plot.height=20)
sum_plot <- covid_final %>%
    ggplot(aes(x=Last_Update, y=Total_Confirmed, color=type, linetype=type)) +
    geom_line() +
    xlab("Updated date") +
    ylab("Total confirmed cases")
sum_plot
```
